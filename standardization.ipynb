```python
scaled_data = df.copy()
```

```python
# z- score method
# scaled_value = value - mean/std
```

```python
# apply the formula 

for col in ['fixed acidity','pH']:
    scaled_data[col] = (scaled_data[col] - scaled_data[col].mean()) / (scaled_data[col].std())
```

```python
sns.distplot(scaled_data['fixed acidity'])
```

```python
sns.distplot(scaled_data['pH'])
```
